```python
glass_data = pd.read_csv('/kaggle/input/glass-identification-data-set/glass.csv')
glass_data.head()
```

```python
len(glass_data)
```

```python
glass_data['Type of glass'].unique()
```

```python
glass_data = glass_data.drop(['Id'],axis=1)
glass_data.head()
```

```python
x = glass_data.iloc[:,:-1] #assigning x variable with all the columns except for the last one
x
```

```python
y=glass_data.iloc[:,-1] #assigning y variable with the output variable
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state= 42)
```

```python
rf_clf = RandomForestClassifier(random_state= 42)
```

```python
rf_clf.fit(x_train, y_train)
```

```python
test_pred = rf_clf.predict(x_test)
test_pred
```

```python
print(classification_report(test_pred, y_test))
```
